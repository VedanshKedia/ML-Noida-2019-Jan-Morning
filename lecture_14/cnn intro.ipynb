```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("../panda.jpeg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
modified = cv2.resize(gray, (50, 50))
plt.imshow(modified, cmap="gray")
```

```python
fil = np.ones([10, 10])
```

```python
output = np.zeros([gray.shape[0] - fil.shape[0] + 1, gray.shape[1] - fil.shape[1] + 1])
```

```python
for i in range(0, gray.shape[0] - fil.shape[0] + 1):
    for j in range(0, gray.shape[1] - fil.shape[1] + 1):
        cut = gray[i:i+fil.shape[0], j:j+fil.shape[1]]
        out = np.sum(cut * fil)
        output[i, j] = out

plt.imshow(output/255, cmap="gray")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../panda.jpeg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap="gray")
```

```python
resized = cv2.resize(gray, (200, 200))
```

```python
plt.imshow(resized, cmap="gray")
```

```python
pool_shape = (4, 4)
```

```python
img_gen = np.zeros([resized.shape[0]//pool_shape[0], resized.shape[1]//pool_shape[1]])
```

```python
img_gen.shape
```

```python
for i in range(0, resized.shape[0], pool_shape[0]):
    for j in range(0, resized.shape[1], pool_shape[1]):
        sl = resized[i:i+pool_shape[0], j:j+pool_shape[1]]
        img_gen[i//pool_shape[0], j//pool_shape[1]] = np.max(sl)
        
        
```

```python
plt.imshow(img_gen, cmap="gray")
```

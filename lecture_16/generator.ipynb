```python
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
panda = image.img_to_array(image.load_img("../panda.jpeg")) / 255
```

```python
plt.imshow(panda)
```

```python
gen = image.ImageDataGenerator(rotation_range=30, zoom_range=.3, shear_range=.2, horizontal_flip=True)
```

```python
for i, batch in enumerate(gen.flow(np.array([panda, panda, panda]), batch_size=2)):
    print(batch.shape)
    for l in range(batch.shape[0]):
        plt.figure(l)
        plt.imshow(batch[l])
    if i == 2:
        break
```

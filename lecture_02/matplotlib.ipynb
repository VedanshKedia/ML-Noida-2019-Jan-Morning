## Plot with x and y

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(100)

plt.plot(x)
```

```python
x = np.linspace(-50, 50, 10) 
y = x ** 2

print(x)
print(y)
plt.plot(x, y, linestyle='none')
plt.scatter(x, y, marker='d')
```

## color and legend

```python
x = np.linspace(-np.pi, np.pi, 1000)

si = np.sin(x)
co = np.cos(x)

plt.plot(x, si, label = "sin", color='red')
plt.plot(x, co, label = "cos", color='yellow')

plt.xlabel("angles")
plt.ylabel("values")

plt.title("My first plot")

plt.legend()

```
